```python

df = pd.read_csv(DATA, sep=';', decimal=',', na_values='', keep_default_na=False)
df.head()
```

```python
df.describe().T
```

```python
df_melted = df.melt(id_vars='EMPRESA',
                    var_name='INDICADOR',
                    value_name='VALOR')
df_melted.head()
```

```python
## Verifica se existem valores NA - Nao aplicavel
df_melted['NA'] = df_melted.VALOR.apply(lambda x: 1 if x=='NA' else 0)
df_melted.loc[df_melted.NA==1, 'VALOR'] = np.NaN

## Verifica se existem valores ND - Nao disponivel
df_melted['ND'] = df_melted.VALOR.apply(lambda x: 1 if x=='ND' else 0)
df_melted.loc[df_melted.ND==1, 'VALOR'] = np.NaN

## Converte os valores para numero
df_melted.VALOR = df_melted.VALOR.str.replace(',','.').astype(np.float32)

## Renomeia os nomes de colunas
df_melted.columns = ['empresa', 'indicador', 'valor', 'na', 'nd']

## Classificao
df_melted.sort_values(['indicador', 'empresa'], inplace=True)

df_melted.head(10)
```

```python
df_pivot = df_melted.pivot(index='empresa', columns='indicador', values='valor')
print(df_pivot.shape)
df_pivot.head()
```
